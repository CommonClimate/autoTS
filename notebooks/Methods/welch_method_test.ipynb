```python
import numpy as np
import pyleoclim as pyleo
from pyleoclim.utils import spectral
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def generate_signal(ratio):
    freqs=[1/20,1/80]
    time=np.arange(5001)
    signals=[]
    for freq in freqs:
        signals.append(np.cos(2*np.pi*freq*time))
    signal=sum(signals)
    sig_var = np.var(signal)
    noise_var = sig_var*ratio#signal is twice the size of noise
    white_noise = np.random.normal(0, np.sqrt(noise_var), size=np.size(signal))
    signal_noise = signal + white_noise
    return time,signal_noise
    
    
```

```python
time,signal = generate_signal(0)
```

```python
plt.plot(time,signal)
```

```python
res_welch =spectral.welch(time,signal,ana_args={'nperseg':len(time)/2})
```

```python
res_periodogram = spectral.periodogram(time,signal)
```
