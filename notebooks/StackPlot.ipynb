```python
time1 = np.arange(150,1000)
time2 = np.arange(100,2000)
time3 = np.arange(100,3000)
time4 = np.arange(500,4000)
f1 = 1/100
y1 = np.cos(2*np.pi*time1*f1)
f2 = 1/200
y2 = np.sin(2*np.pi*time2*f2)
f3 = 1/200
y3 = np.cos(2*np.pi*time3*f3)
f4 = 1/300
y4 = np.cos(2*np.pi*time4*f4)
```

```python
ts1 = pyleo.Series(time=time1,value=y1)
ts2 = pyleo.Series(time=time2,value=y2)
ts3 = pyleo.Series(time=time3,value=y3)
ts4 = pyleo.Series(time=time4,value=y4)
```

```python
series=[ts1,ts2,ts3,ts4]
```
