```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
time1 = np.arange(1000)
time2 = np.arange(2000)
time3 = np.arange(3000)
```

```python
f1 = 1/100
y1 = np.cos(2*np.pi*time1*f1)
f2 = 1/200
y2 = np.sin(2*np.pi*time2*f2)
f3 = 1/200
y3 = np.cos(2*np.pi*time3*f3)
```

```python
def stackplot(**data):
    x_values = data['x']
    y_values = data ['y']
    fig,ax = plt.subplots()
  
    
    for x,y in zip(x_values,y_values):
        ax.plot(x,y)
    plt.show()
```

```python
d = {'x':[time1,time2,time3],'y':[y1,y2,y3]}
```

```python
stackplot(**d)
```
